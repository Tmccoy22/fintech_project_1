```python
import os
import requests
import json
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
from MCForecastTools import MCSimulation
%matplotlib inline
```

```python
load_dotenv()
```

```python
# Set the variables for the Alpaca API and secret keys
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

display(type(alpaca_secret_key))
display(type(alpaca_api_key))
```

```python
# Create the Alpaca tradeapi.REST object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = "v2")
```

```python
# Set the tickers for both the bond and stock portion of the portfolio
tickers = ["PM","TAP", "CZR","RGR","SAM"]

# Set timeframe to 1Day
timeframe = "1D"

# Format current date as ISO format
# Set both the start and end date at the date of your prior weekday 
# This will give you the closing price of the previous trading day
# Alternatively you can use a start and end date of 2020-08-07
start_date = pd.Timestamp("2017-01-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2022-12-31", tz="America/New_York").isoformat()
```

```python
# Use the Alpaca get_bars function to get current closing prices the portfolio
# Be sure to set the `df` property after the function to format the response object as a DataFrame
prices_df = alpaca.get_bars(
    tickers,
    timeframe,
    start=start_date,
    end=end_date,
).df

# Reorganize the DataFrame
# Separate ticker data
PM = prices_df[prices_df['symbol']=='PM'].drop('symbol', axis=1)
TAP = prices_df[prices_df['symbol']=='TAP'].drop('symbol', axis=1)
CZR = prices_df[prices_df['symbol']=='CZR'].drop('symbol', axis=1)
RGR = prices_df[prices_df['symbol']=='RGR'].drop('symbol', axis=1)
SAM = prices_df[prices_df['symbol']=='SAM'].drop('symbol', axis=1)


# Concatenate the ticker DataFrames
prices_df = pd.concat([PM, TAP, CZR, RGR, SAM], axis=1, keys=["PM","TAP", "CZR","RGR","SAM"])

# Review the first 5 rows of the Alpaca DataFrame
prices_df.head()
```

```python
MC_thirty_year = MCSimulation(
    portfolio_data = prices_df,
    weights = [.20,.20,.20,.20,.20],
    num_simulation = 500,
    num_trading_days = 252*30
)

# Review the simulation input data
MC_thirty_year.portfolio_data.head()
```

```python
MC_thirty_year.calc_cumulative_return()
```

```python
line_plot = MC_thirty_year.plot_simulation()
```

```python
distribution_plot = MC_thirty_year.plot_distribution()
```

```python
MC_table = MC_thirty_year.summarize_cumulative_return()
```

```python
print(MC_table)
```

```python
investment = 10000
```

```python
print(f"We will invest ${investment} spread evenly over the five vice stocks PM, TAP, CZR, RGR , SAM.")
```

```python
lower_cumulative_return = round(MC_table[8]*investment,2)
upper_cumulative_return = round(MC_table[9]*investment,2)

print(f"There is a 95% chance that the portfolio"
      f" will end in 30 years within in the range of"
      f" ${lower_cumulative_return} and ${upper_cumulative_return}.")
```

```python
#compair the portfolio to SPY index benchmark
# Set the variables for the Alpaca API and secret keys
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

display(type(alpaca_secret_key))
display(type(alpaca_api_key))
```

```python
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = "v2")
```

```python
# Set the tickers for both the bond and stock portion of the portfolio
ticker = ["SPY"]

# Set timeframe to 1Day
timeframe = "1D"

# Format current date as ISO format
# Set both the start and end date at the date of your prior weekday 
# This will give you the closing price of the previous trading day
# Alternatively you can use a start and end date of 2020-08-07
start_date = pd.Timestamp("2017-01-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2022-12-31", tz="America/New_York").isoformat()
```

```python
# Use the Alpaca get_bars function to get current closing prices the portfolio
# Be sure to set the `df` property after the function to format the response object as a DataFrame
spy_df = alpaca.get_bars(
    ticker,
    timeframe,
    start=start_date,
    end=end_date,
).df

# Reorganize the DataFrame
# Separate ticker data
SPY = spy_df[spy_df['symbol']=='SPY'].drop(columns = 'symbol', axis=1)
spy_df = pd.concat([SPY], axis=1, keys=["SPY"])

# Review the first 5 rows of the Alpaca DataFrame
spy_df.head()
```

```python
MC_thirty_year = MCSimulation(
    portfolio_data = spy_df,
    weights = [1],
    num_simulation = 500,
    num_trading_days = 252 * 30
)

# Review the simulation input data
MC_thirty_year.portfolio_data.head()
```

```python
MC_thirty_year.calc_cumulative_return()
```

```python
line_plot = MC_thirty_year.plot_simulation()
```

```python
distribution_plot = MC_thirty_year.plot_distribution()
```

```python
MC_table = MC_thirty_year.summarize_cumulative_return()
print(MC_table)
```
